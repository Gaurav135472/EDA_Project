Clean up data

```python
# Only want USA races , 50k or 50Mi 2020
df[df['Event distance/length'] == '50km']
```

```python
df[df['Event distance/length'] == '50mi']
```

```python
## COmbine 50k or 50mi with isin
df[df['Event distance/length'].isin(['50km', '50mi'])]
```

```python
df[(df['Event distance/length'].isin(['50km', '50mi'])) & (df['Year of event'] == 2020) ]
```

```python
df[df['Event name'] == 'Everglades 50 Mile Ultra Run (USA)']['Event name'].str.split("(").str.get(1).str.split(")").str.get(0)
```

```python
df[df['Event name'].str.split("(").str.get(1).str.split(")").str.get(0) == "USA"]
```

```python
# Combine all the filters together
df2 = df[(df['Event distance/length'].isin(['50km', '50mi'])) & (df['Year of event'] == 2020) & (df['Event name'].str.split("(").str.get(1).str.split(")").str.get(0) == "USA") ]
```

```python
df2['Event name'] = df2['Event name'].str.split("(").str.get(0)
```

```python
# Clean up athlete age
df2['athlete_age'] = 2020.00 - df2['Athlete year of birth'].astype(float)
```

```python
df2.head()
```

```python
# Remove h from athlere performance 

df2['Athlete performance'] = df2['Athlete performance'].str.split(" ").str.get(0)
```

```python
df2.head(5)
```

```python
# Drop columns : athlere clubs , Athlete Country, Athlete year of birth, Athlete category

df2.drop(columns=['Athlete club', 'Athlete country', 'Athlete year of birth', 'Athlete age category'], inplace=True)
```

```python
df2.head()
```

```python
# Clean up all null values

df2.isna().sum()
```

```python
df2[df2['athlete_age'].isna() == 1]
```

```python
df2 = df2.dropna()
```

```python
# Check dupes

df2[df2.duplicated() == True]
```

```python
df2.reset_index(drop=True)
```

```python
# fix types

df2.dtypes
```

```python
df2['athlete_age'] = df2['athlete_age'].astype(int)
df2['Athlete average speed'] = df2['Athlete average speed'].astype(float)
```

```python
# Rename columns 

df2 = df2.rename(columns={
    'Year of event': 'year',
    'Event dates': 'race_day',
    'Event name': 'race_name',
    'Event distance/length': 'race_length',
    'Event number of finishers': 'race_number_of_finisher', 
    'Athlete performance': 'performance',
    'Athlete gender': 'gender' ,
    'Athlete average speed': 'speed',
    'Athlete ID' : 'id',
    'athlete_age': 'age',
})
```

```python
df2.columns
```

```python
# reorder columns

df3 = df2[['race_day', 'race_name', 'race_length', 'race_number_of_finisher', 'id','gender', 'performance', 'speed', 'age']]
```
